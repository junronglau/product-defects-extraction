### Plot multi chart and see which category has been spiking recently

```python

fig,ax = plt.subplots(figsize=(20,10))

for name in df['category'].unique():
    alpha=0.2
    if name in ['colour issues', 'consistency issues', 'infection issues', 'broken products', 'unsealed packages']:
        alpha = 1
    mascara_df = df[df['category']==str(name)]
    converted = mascara_df['ASIN'][mascara_df['converted_date']>'2011-01-01'].groupby([mascara_df['converted_date'].dt.year]).agg('count')    
    ax.plot(converted.index, converted.values, label=name, linewidth=7, alpha=alpha)

ax.legend(loc='best', fontsize=20)
plt.show()
```
